```python
import ColabTurtle.Turtle as honu
honu.initializeTurtle()
honu.right(90) #honu.rt()
honu.forward(100) #honu.fd()
honu.left(90) #honu.lt()
honu.backward(100) #honu.bk()
```

Coordinate System![Coordinate System]()

```python
#Honu Tests the Coordinate System.
honu.home()
honu.goto(100,100)
honu.goto(100,50)
honu.goto(50,50)
honu.goto(50,100)

#Honu makes a Square!
honu.home()
for i in range(4):
  honu.fd(100)
  honu.rt(90)
```

```python
#Honu Changes the Background.
#honu.circle(60) A COMMAND THAT EXISTS IN TURTLE BUT NOT COLABTURTLE
#honu.dot(20) A COMMAND THAT EXISTS IN TURTLE BUT NOT COLABTURTLE
honu.bgcolor("red")
honu.bgcolor("green")
honu.bgcolor("blue")
#honu.title("Honu's Program") A COMMAND THAT EXISTS IN TURTLE BUT NOT COLABTURTLE
```

```python
#Honu Becomes a Green Circle :,(
honu.initializeTurtle()
honu.shape("circle")
honu.pensize(4)
honu.fd(100)

honu.pencolor("green")
#honu.color("white", "green") A COMMAND THAT EXISTS IN TURTLE BUT NOT COLABTURTLE
```

```python
honu.initializeTurtle()
#honu.begin_fill() A COMMAND THAT EXISTS IN TURTLE BUT NOT COLABTURTLE
honu.fd(100)
honu.lt(120)
honu.fd(100)
honu.lt(120)
honu.fd(100)
#honu.end_fill() A COMMAND THAT EXISTS IN TURTLE BUT NOT COLABTURTLE

#Honu becomes speedy.
honu.speed(1)
honu.fd(100)
honu.speed(10)
honu.fd(100)
```

```python
honu.initializeTurtle()
#honu.pen(pencolor="purple", fillcolor="orange", pensize=10, speed=9) A COMMAND THAT EXISTS IN TURTLE BUT NOT COLABTURTLE

#Honu loses and gets back some pen ink.
honu.fd(100)
honu.rt(90)
honu.penup()
honu.fd(100)
honu.rt(90)
honu.pendown()
honu.fd(100)
honu.rt(90)
honu.penup()
honu.fd(100)
honu.pendown()
#honu.undo() A COMMAND THAT EXISTS IN TURTLE BUT NOT COLABTURTLE
honu.clear()
#honu.reset() A COMMAND THAT EXISTS IN TURTLE BUT NOT COLABTURTLE
```
